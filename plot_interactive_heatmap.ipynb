```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import colormaps as cm
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/binouyang/Desktop/eta.csv')
```

```python
metals = [
    'Sc','Ti','V','Cr','Mn','Fe','Co','Ni','Cu','Zn',
    'Y','Zr','Nb','Mo','Ru','Rh','Pd','Ag','Cd',
    'La','Hf','Ta','W','Re','Os','Ir','Pt','Au'
]

n_metal = len(metals)

HER_OP = np.nan*np.ones([n_metal,5])
ORR_OP = np.nan*np.ones([n_metal,5])
metal_names = np.empty([n_metal,5], dtype='<U256')
images_HER, images_ORR = np.empty([n_metal,5], dtype='<U256'), np.empty([n_metal,5], dtype='<U256')
```

```python
ls_dict = {'N4':4,'C1N3':3,'C2N2':2,'C3N1':1,'C4':0}
ls_tag = {0:'0N',1:'1N',2:'2N',3:'3N',4:'4N'}
ordered_metals = []

for index, row in df.iterrows():
    row_items = list(row)
    SAC_name = row_items[0]
    metal, ls = SAC_name.split('@')
    
    if metal == 'Ce':
        continue
    
    col_ind = ls_dict[ls]
    row_ind = metals.index(metal)
    image_name_HER = f"./predicted_performance/FreeEnergy/HER/{col_ind}N_{metal}.jpg"
    image_name_ORR = f"./predicted_performance/FreeEnergy/ORR/{col_ind}N_{metal}.jpg"
    ordered_metals.append(metal)
    
    HER_OP[row_ind, col_ind] = row_items[1]
    ORR_OP[row_ind, col_ind] = row_items[2]
    metal_names[row_ind, col_ind] = metal
    images_HER[row_ind, col_ind] = image_name_HER
    images_ORR[row_ind, col_ind] = image_name_ORR
    
    print(metal, ls, row_ind, col_ind, image_name_HER, image_name_ORR)

print(HER_OP.shape)
print(ORR_OP.shape)
```

### Plot the HER figure

```python
# Create a grid of coordinates for the heatmap
print(HER_OP.shape)
x_range = list(range(HER_OP.shape[0]))
y_range = list(range(HER_OP.shape[1]))

xx, yy = np.meshgrid(y_range, x_range)

# Flatten the grid and data for the ColumnDataSource
x_coords = yy.flatten()
y_coords = xx.flatten()
values = HER_OP.flatten()
imgs = images_HER.flatten()
M_names = metal_names.flatten()
print(x_coords,M_names)
print(metal_names.shape)
print(HER_OP)

width, height = images_HER.shape
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import HoverTool, ColumnDataSource, ColorBar
from bokeh.transform import linear_cmap
from bokeh.io import push_notebook
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from bokeh.models import Div
from bokeh.layouts import column, row
from bokeh.embed import components

# Create a ColumnDataSource
print(x_coords.shape)
source = ColumnDataSource(data=dict(
    x=x_coords,
    y=y_coords,
    values=values,
    imgs=imgs,
    M_names=M_names
))

TOOLTIPS = """
    <div>
        <div style="display: flex; align-items: center; justify-content: center; height: 100%; width: 100%;">
            <img
                src="@imgs" height="250" alt="@imgs" width="366"
                style="margin: 300px 15px 15px 50px;"
                border="2"
            ></img>
        </div>
        <div style="text-align: center;">
            <span style="font-size: 17px;"><span style="font-weight: bold;">System: </span><span>@{M_names}@@{y}N</span><br>
            <span style="font-size: 17px;"><span style="font-weight: bold;">Overpotential: </span><span>@values{0.00} V</span>
        </div>
    </div>
"""

# Create a new plot
colormap = cm.coolwarm
coolwarm_palette = [mcolors.rgb2hex(m) for m in colormap(np.arange(colormap.N))]
mapper = linear_cmap(field_name='values', palette=coolwarm_palette, low=min(values), high=max(values))

p = figure(width=1000, height=200, x_range=(-0.5, width-0.5), y_range=(-0.5, height-0.5))
p.rect(x="x", y="y", width=1, height=1, source=source, line_color="white", fill_color=mapper)

# Add hover tool
hover = HoverTool()
hover.tooltips = TOOLTIPS
p.add_tools(hover)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Change tick
p.yaxis.ticker = list(range(5))
p.xaxis.ticker = list(range(len(ordered_metals)))

p.yaxis.major_label_overrides = {i:ls_tag[i] for i in range(5)}
p.xaxis.major_label_overrides = {i:ordered_metals[i] for i in range(len(ordered_metals))}

output_file("heatmap_HER.html",title='HER Performance map')
show(p)

```
